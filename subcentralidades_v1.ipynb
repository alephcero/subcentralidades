```python
pois_wrapper = []
for departamento_nombre in departamentos_nombres:
    print(departamento_nombre)
    departamento_osm = departamentos_config[departamento_nombre]['nombre_osm']    
    query = ox.geometries_from_place(departamento_osm, {'amenity':True, 'shop':True}) 
    amenities = query.reindex(columns = ['unique_id','amenity','geometry']).dropna(subset=['amenity']).rename(columns = {'amenity':'type'})
    amenities = amenities.loc[~amenities['type'].isin(amenities_eliminar),:] 
    
    shops = query.reindex(columns = ['unique_id','shop','geometry']).dropna(subset=['shop']).rename(columns = {'shop':'type'})
    shops = shops.loc[~shops['type'].isin(shops_eliminar),:] 

    
    pois = pd.concat([amenities,shops])

    
    poligono = ox.geocode_to_gdf(departamento_osm)
    poligono.to_file('clusters/'+departamento_nombre+'_poligono')

    pois = pois.to_crs(epsg=5348)
    pois['geometry'] = pois.geometry.centroid
    pois['id'] = range(len(pois))
    pois['x'] = pois.geometry.x
    pois['y'] = pois.geometry.y
    print(pois['type'].value_counts())
    
    # cluster
    X = pois.loc[:,['x','y']].values
    dbscan_lineas = DBSCAN(eps=departamentos_config[departamento_nombre]['eps'],
                           min_samples=departamentos_config[departamento_nombre]['min_samples']).fit(X)
    pois['cluster']=dbscan_lineas.labels_
    
    etiquetas_clusters = pois.cluster.value_counts()[pois.cluster.value_counts().index>-1].index
    etiquetas_por_tamanio = {k:v for k,v in zip(etiquetas_clusters,range(len(etiquetas_clusters)))}
    pois.cluster = pois.cluster.replace(etiquetas_por_tamanio)

    pois['municipio'] = departamento_nombre
    print(pois['cluster'].value_counts())
    
    pois_wrapper.append(pois)
    pois.to_file('clusters/'+departamento_nombre)
```

```python
final = pd.concat(pois_wrapper)
final.to_file('clusters/final.geojson',driver='GeoJSON')
```
