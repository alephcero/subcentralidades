```python
amenities_wrapper = []
for departamento_nombre in departamentos_nombres:
    print(departamento_nombre)
    departamento_osm = departamentos_config[departamento_nombre]['nombre_osm']
    amenities = ox.geometries_from_place(departamento_osm, {'amenity':True}) 
    amenities = amenities.reindex(columns = ['amenity','geometry'])
    poligono = ox.geocode_to_gdf(departamento_osm)
    poligono.to_file('clusters/'+departamento_nombre+'_poligono')

    amenities = amenities.to_crs(epsg=5348)
    amenities['geometry'] = amenities.geometry.centroid
    amenities['id'] = range(len(amenities))
    amenities['x'] = amenities.geometry.x
    amenities['y'] = amenities.geometry.y
    print(amenities.amenity.value_counts())
    
    # cluster
    X = amenities.loc[:,['x','y']].values
    dbscan_lineas = DBSCAN(eps=200,min_samples=5).fit(X)
    amenities['cluster']=dbscan_lineas.labels_
    amenities['municipio'] = departamento_nombre
    print(amenities['cluster'].value_counts())
    amenities_wrapper.append(amenities)
    amenities.to_file('clusters/'+departamento_nombre)
    
```

```python
import pandas as pd
```

```python
final = pd.concat(amenities_wrapper)
```

```python
final.to_file('clusters/final.geojson',driver='GeoJSON')
```
